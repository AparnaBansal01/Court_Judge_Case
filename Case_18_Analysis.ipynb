```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import plotly.express as px
%matplotlib inline 
import seaborn as sns
from datetime import datetime
import warnings                  
warnings.filterwarnings("ignore")
```

```python
#case_10 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IIITH/cases_2010.csv')
case_18 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IIITH/cases_2018.csv')
purp_name_key = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IIITH/purpose_name_key.csv")
type_name_key = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IIITH/type_name_key.csv")
disp_name_key = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IIITH/disp_name_key.csv')
court_key = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/IIITH/cases_court_key.csv")
```

```python
date_format = "%Y-%m-%d"
duration = []
```

```python
current_case_18 = case_18[case_18['date_of_decision'].isna()]
```

```python
current_case_18 = pd.merge(current_case_18, court_key, how = 'inner', on =['state_code', 'dist_code', 'court_no','year'])
```

```python
fig=plt.figure(figsize=(15,7))
plt.xticks(rotation=90,fontsize=12)
plt.title('Number of cases state wise',fontdict={'fontsize':13})
plots = sns.barplot(y = 'type_name',  x='state_name', data=current_case_18.groupby('state_name').count().reset_index(),palette='autumn');
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=9, xytext=(0, 8),
                   textcoords='offset points')
 
plt.ylabel('Cases',fontdict={'fontsize':12})
plt.xlabel('State',fontdict={'fontsize':12})
sns.despine(bottom = True, left = True);
plt.show()
```

```python
case_18.isnull().sum().sum()
```

```python
case_18['date_of_decision'].isnull().sum()
```

```python
case_18 = case_18.dropna()
case_18 = case_18.reset_index(drop = True)
```

```python
for i in case_18.index:
  duration.append((datetime.strptime(str(case_18['date_of_decision'][i]), date_format) - datetime.strptime(str(case_18['date_of_filing'][i]), date_format)).days)
```

```python
case_18['Duration'] = duration
```

```python
case_18.head()
```

```python
case_18 = case_18.drop(['ddl_case_id','cino', 'date_of_filing','date_of_decision', 'date_first_list', 'date_last_list', 'date_next_list'], axis =1)
```

```python
purp_name_key.head()
```

```python
disp_name_key.head()
```

```python
type_name_key.head()
```

```python
type_u = type_name_key.drop('count', axis = 1)
case_18 = pd.merge(case_18, type_u, how = 'inner', on =['type_name', 'year'])
```

```python
disp_u = disp_name_key.drop('count', axis = 1)
case_18 = pd.merge(case_18, disp_u, how = 'inner', on =['disp_name', 'year'])
```

```python
purp_u = purp_name_key.drop('count', axis = 1)
case_18 = pd.merge(case_18, purp_u, how = 'inner', on =['purpose_name', 'year'])
```

```python
case_18 = pd.merge(case_18, court_key, how = 'inner', on =['state_code', 'dist_code', 'court_no','year'])
```

```python
case_18.to_csv('/content/drive/MyDrive/Colab Notebooks/IIITH/merged_cases_2018.csv')
```
